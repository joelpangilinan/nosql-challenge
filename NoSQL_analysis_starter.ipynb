### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'$match': {'scores.Hygiene': {'$eq': 20}}}
pipeline = [query]
results = list(establishments.aggregate(pipeline))
pprint(results)
```

```python
# Use count_documents to display the number of documents in the result
count_hygiene = establishments.count_documents({'scores.Hygiene': {'$eq': 20}})
count_hygiene
```

```python
# Display the first document in the results using pprint
pprint(results[0:1])
```

```python
# Convert the result to a Pandas DataFrame
establishments_df = pd.json_normalize(results)
establishments_df.head()
```

```python
# Display the number of rows in the DataFrame
len(establishments_df)
```

```python
# Display the first 10 rows of the DataFrame
establishments_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'$match': {'LocalAuthorityName': {'$regex': 'London'},
                    'RatingValue': {'$gte': '4'}}}
pipeline = [query]
results = list(establishments.aggregate(pipeline))
pprint(results)
```

```python
# Use count_documents to display the number of documents in the result
count_local_authority = establishments.count_documents({'LocalAuthorityName': {'$regex': 'London'},
                                                       'RatingValue': {'$gte': '4'}})
count_local_authority
```

```python
# Display the first document in the results using pprint
pprint(results[0:1])
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.json_normalize(results)
london_df.head()
```

```python
# Display the number of rows in the DataFrame
len(london_df)
```

```python
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Based on the information provided on the challenge,
# the latitude is 51.49014200 and longitude is 0.08384000 for Penang Flavour
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

latitude_lte = latitude+degree_search
latitude_gte = latitude-degree_search

longitude_lte = longitude+degree_search
longitude_gte = longitude-degree_search

# Create the query
query = {'$match': {'RatingValue': {'$eq': '5'},
                   'geocode.latitude': {'$gte': latitude_gte, '$lte': latitude_lte},
                   'geocode.longitude': {'$gte': longitude_gte, '$lte': longitude_lte}}}

# Sort by hygiene score
sort_values = {'$sort': {'scores.Hygiene': -1}}

# Create the pipeline and limit the results to 5
pipeline = [query, sort_values, {'$limit': 5}]

# Print the results
results = list(establishments.aggregate(pipeline))
pprint(results)
```

```python
# Convert result to Pandas DataFrame
top_five_df = pd.json_normalize(results)
top_five_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': {'LocalAuthorityName': '$LocalAuthorityName'}, 'count': {'$sum': 1}}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': { 'count': -1 }}

# Put the pipeline together
pipeline = [match_query, group_query, sort_values]

# Print the number of documents in the result
results = list(establishments.aggregate(pipeline))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
local_authority_hygiene_df = pd.json_normalize(results)
```

```python
# Display the number of rows in the DataFrame
len(local_authority_hygiene_df)
```

```python
# Display the first 10 rows of the DataFrame
local_authority_hygiene_df.head(10)
```

```python
# Rename the columns
local_authority_hygiene_df = local_authority_hygiene_df.rename(columns={"_id.LocalAuthorityName":"_id"})
local_authority_hygiene_df.head(10)
```

```python
# Reorder the columns
local_authority_hygiene_df = local_authority_hygiene_df[["_id", "count"]]
local_authority_hygiene_df.head(10)
```
